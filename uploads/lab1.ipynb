```python
# 1. Load a gray scale image
# 2. Read the intensities of the image and display in an array
# 3. Add any constant value with each pixel and display the final image
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_path = 'happy_dog.jpg'
image = plt.imread(image_path)
```

```python
image = Image.open(image_path).convert('L')
plt.imshow(image, cmap='gray')
plt.title("Image")
plt.axis('off')
plt.show()
```

```python
# flattened_image = image.flatten()
intensities = np.array(image)
print("Original Intensities:", intensities)
```

```python
padding = 1  # Number of pixels to pad on each side
padded_image = np.pad(image, pad_width=padding, mode='constant', constant_values=0)

# Display the padded image
plt.imshow(padded_image, cmap='gray')
plt.title("Padded Image")
plt.axis('off')
plt.show()
```

```python
constant_value = 20  # Replace with your desired constant value

# Adding the constant value to the intensities
final_intensities = intensities + constant_value
```

```python
plt.imshow(final_intensities, cmap='gray')
plt.title("Final Image")
plt.axis('off')
plt.show()
```

```python
intensities = np.array(image)
height, width = intensities.shape
height
```

```python
new_image_np = intensities
for y in range(1,width-2):
    for x in range(1,height-2):
        new_in = (intensities[x-1][y] + intensities[x+1][y] + intensities[x][y-1] + intensities[x][y+1] )
        new_image_np[x][y] = (new_in/4) % 255
new_image_np
```

```python
plt.imshow(new_image_np, cmap='gray')
plt.title("Final Image")
plt.axis('off')
plt.show()
```
